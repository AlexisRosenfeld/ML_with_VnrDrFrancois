```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def test(ax, l, label):
    # Paramètres
    n_individus = 30  # Nombre d'individus par classe
    n_classes = 2     # Nombre de classes
    n_dimensions = 2  # Deux dimensions pour chaque individu
    data = []
    for i in range(n_classes):
        class_data = np.random.uniform(low=l[i][0], high=l[i][1], size=(n_individus, n_dimensions))
        class_labels = np.full((n_individus, 1), i)  # Etiquette de classe (0 ou 1)
        class_with_labels = np.hstack((class_data, class_labels))  # Ajouter les étiquettes
        data.append(class_with_labels)
    
    # Convertir les données en un DataFrame pour une manipulation plus facile
    df = pd.DataFrame(np.vstack(data), columns=[f'Dimension_{j+1}' for j in range(n_dimensions)] + ['Label'])
    
    # Séparer les données par classe pour les tracer
    class_0 = df[df['Label'] == 0]
    class_1 = df[df['Label'] == 1]
    
    # Tracer les données pour chaque classe sans légende
    ax.scatter(class_0['Dimension_1'], class_0['Dimension_2'], color='blue')
    ax.scatter(class_1['Dimension_1'], class_1['Dimension_2'], color='red')
    ax.set_title(f'{label}')
    ax.set_xlabel('Dimension 1')
    ax.set_ylabel('Dimension 2')
    ax.grid(True)

# Créer une fenêtre avec deux sous-plots de plus petite taille
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 4))

# Deuxième distribution avec le titre "Close Distribution"
distrib2 = [[-1, 0], [-0.5, 0.5]]
test(ax2, distrib2, 'Close Distribution')

# Première distribution avec le titre "Disjoint Distribution"
distrib1 = [[-1, 0], [-0.2, 0.8]]
test(ax1, distrib1, 'Disjoint Distribution')

# Ajuster l'espacement entre les sous-plots
plt.tight_layout()

# Afficher le plot avec les deux sous-plots
plt.show()
```
